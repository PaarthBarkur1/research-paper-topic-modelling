```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
#from pywsd.utils import lemmatize_sentence
import spacy

def clean(doc):
    sent = "".join(list(i for i in str(doc) if ((i.isalpha())|(i ==' '))))
    return sent
                   

def preprocess(df):
    nlp = spacy.load('en_core_web_sm')
    
    words=[]
    sw = set(stopwords.words('english'))
    for i in df['ABSTRACT']:
        sent=clean(i)
        doc=nlp(sent)
        words.append([token.lemma_ for token in doc if token.lemma_ not in sw])
    df['Words'] = words
    
```

```python
from gensim.models import Word2Vec
wv=Word2Vec(train['Words'],vector_size=100, window=5, min_count=2, sg=0)
```

```python
def handler(doc, wv):
    col = [i for i in doc if i in wv.wv.index_to_key]
    if len(col) > 0:
        return col
    else:
        return ['unknown']
```

```python
train['vectors']=[wv.wv[handler(i,wv)] for i in train['Words']]
train.head()
```

```python
train['vectors'][0]
```

```python
paracol=['Computer Science','Mathematics','Physics','Statistics','vectors']
```

```python
actualtrain=train[paracol]
y=train[tarcols]
```

```python
def summarizer(df):
    flag=0
    for i in df['vectors']:
        if flag==0:
            x=np.mean(i,axis=0)
            flag=1
        else:
            x=np.vstack([x,np.mean(i,axis=0)])
    #df.drop('vectors',axis=1,inplace=True)
    return x
```

```python
x=summarizer(actualtrain)
```

```python
x.shape
```

```python
def combiner(df,x):
    a,b=x.shape
    y=np.zeros((a,b+4))
    for i in range(len(df)):
        y[i]=np.hstack([x[i],(np.array([df.iloc[i]['Computer Science'],df.iloc[i]['Mathematics'],df.iloc[i]['Physics'],df.iloc[i]['Statistics']]))])
    #df = df.drop(['Computer Science','Mathematics','Physics','Statistics','Vectors'],axis=1)
    #df['Para']=(temp)
    return y
```

```python
import warnings
warnings.filterwarnings('ignore')
x= combiner(actualtrain,x)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

clf=OneVsRestClassifier(KNeighborsClassifier())
clf.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import accuracy_score, f1_score,classification_report
ypred=clf.predict(xtest)
f1_score(ytest,ypred,average='micro')
```

```python
from sklearn.svm import SVC
clf2=OneVsRestClassifier(SVC())
clf2.fit(xtrain,ytrain)
ypred2=clf2.predict(xtest)
f1_score(ytest,ypred2,average='micro')
```

```python
from sklearn.linear_model import LogisticRegression
clf3=OneVsRestClassifier(LogisticRegression())
clf3.fit(xtrain,ytrain)
ypred3=clf3.predict(xtest)
f1_score(ytest,ypred3,average='micro')
```

```python
import xgboost as xgb


xgb_train = xgb.DMatrix(xtrain, ytrain, enable_categorical=True)
xgb_test = xgb.DMatrix(xtest, ytest, enable_categorical=True)
n=50
params = {
    'max_depth': 3,
    'learning_rate': 0.01,
    
}
 
clf4 = xgb.train(params=params,dtrain=xgb_train,num_boost_round=n)


ypred4 = clf4.predict(xgb_test)
```

```python
def sigmoid(preds):
    flag=0
    for i in preds:
        maxi=i.max()
        i = np.where(i == maxi,1,0)
        if flag ==0:
            arr=i
            flag = 1
        else:
            arr=np.vstack((arr,i))
    return arr
```

```python
def vectorizer(df,wv):
    df['vectors']=[wv.wv[handler(i,wv)] for i in df['Words']]
    return df
```

```python
def pipeline(df,clf,wv):
    preprocess(df)
    df = vectorizer(df,wv)
    x= summarizer(df)
    x= combiner(df,x)
    ypred=clf.predict(x)
    return ypred
```

```python
pipeline(test,clf,wv)
```
